```python
import newspaper
#Languages Supported by Newspaper Module
#newspaper.languages()
my_article = Article(url, language="en")
my_article.download()
# Print Html Only
#print(my_article.html)
```

```python
#Parsing The Article - Get the title, author, publishing date and details about the article
my_article.parse()
#Extract The Title
print('Title:', my_article.title)
```

```python
#Extract The Authors
print('Authors:',my_article.authors)
```

```python
#Publishing Date
print('Publishing date:', my_article.publish_date)
```

```python
import nltk
nltk.download('punkt')
#Downloaded nltk in order to use NLP for Keywords and Summaries
```

```python
#Article NLP
my_article.nlp()
```

```python
#Extract Summary
print('Summary:', my_article.summary)
```

```python
#Extract Keywords
print('Keywords:', my_article.keywords)
```
